```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "countries_of_the_world.csv"
countries_df = pd.read_csv(csv_file)
countries_df.head()
```

```python
csv_file = "efw_cc.csv"
happy_df = pd.read_csv(csv_file)
happy_df.head(10)
```

```python
clean_country = countries_df[["Country", "Region", "Population","Pop. Density (per sq. mi.)","Infant mortality (per 1000 births)","GDP ($ per capita)","Literacy (%)","Birthrate","Deathrate"]]
clean_country.count()
final_country = clean_country.dropna(how='any')
final_country.dtypes
```

```python
new_happy_df= happy_df[['year', 'ISO_code', 'countries', 'ECONOMIC FREEDOM', 'rank']].copy()
new_happy_df.head()
#new_happy_df.count()
```

```python
# drop null rows
no_null_new_happy_df = new_happy_df.dropna(how='any')
```

```python
# verify counts
no_null_new_happy_df.count()
```

```python
# Filter the data so that only 2016 year are in a DataFrame
no_null_new_happy_df_2016 = no_null_new_happy_df.loc[(no_null_new_happy_df["year"] == 2016) ]
no_null_new_happy_df_2016.head()
```

```python
# verify counts
no_null_new_happy_df_2016.count()
```

```python
# Convert the column name
new_happy_df_2016 = no_null_new_happy_df_2016.rename(
    columns={"year": "Year", "ISO_code": "ISO code", "countries": "Country", "ECONOMIC FREEDOM": "Economic Freedom", "rank": "Ranking"})
new_happy_df_2016.head()
```
