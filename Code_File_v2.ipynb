```python
# Reformat data
new_pop_density = []
new_infant_mortality = []
new_literacy = []
new_birthrate = []
new_deathrate = []
for row in country_demo['Pop_Density']:
    s = row.replace(',', '.')
    new_pop_density.append(s)

for row in country_demo['Infant_mortality']:
    ss = row.replace(',', '.')
    new_infant_mortality.append(ss)
    
for row in country_demo['Literacy']:
    sss = row.replace(',', '.')
    new_literacy.append(sss)

for row in country_demo['Birthrate']:
    ssss = row.replace(',', '.')
    new_birthrate.append(ssss)
    
for row in country_demo['Deathrate']:
    sssss = row.replace(',', '.')
    new_deathrate.append(sssss)

se = pd.Series(new_pop_density)
see = pd.Series(new_infant_mortality)
seee = pd.Series(new_literacy)
seeee = pd.Series(new_birthrate)
seeeee = pd.Series(new_deathrate)
```

```python
# Append data into a new DataFrame
items_pd_2 = pd.DataFrame(se, columns=[
                        "New_Pop_Density"])
#items_pd_2.head()
items_pd_3 = pd.DataFrame(see, columns=[
                        "New_Infant_mortality"])

items_pd_4 = pd.DataFrame(seee, columns=[
                        "New_Literacy"])


items_pd_5 = pd.DataFrame(seee, columns=[
                        "New_Birthrate"])

items_pd_6 = pd.DataFrame(seee, columns=[
                        "New_Deathrate"])
```

```python
# Convert data into numeric
items_pd_2['New_Pop_Density'] = pd.to_numeric(items_pd_2['New_Pop_Density'], errors='coerce')
items_pd_3['New_Infant_mortality'] = pd.to_numeric(items_pd_3['New_Infant_mortality'], errors='coerce')
items_pd_4['New_Literacy'] = pd.to_numeric(items_pd_4['New_Literacy'], errors='coerce')
items_pd_5['New_Birthrate'] = pd.to_numeric(items_pd_5['New_Birthrate'], errors='coerce')
items_pd_6['New_Deathrate'] = pd.to_numeric(items_pd_6['New_Deathrate'], errors='coerce')
#items_pd_2.head()
```
